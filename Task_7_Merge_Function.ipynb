```python
import pandas as pd
```

```python
left = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Shanu', 'Amar', 'Santosh', 'Rahul', 'Jayesh'],
   'Lang':['English','Hindi','Marathi','Telugu','Gujarathi']})
right = pd.DataFrame(
   {'id':[1,2,3,4,5],
   'Name': ['John', 'Dinesh', 'Ranbir', 'Rahul', 'Jayesh'],
   'Lang':['English','Hindi','Tamil','Bangali','Malyalam']})

print (pd.merge(left,right,on='id'))
```

```python
st_18 = pd.read_csv('D:\John Sir\Pandas\Sales Transactions-2018.csv')
st_18.head()
```

```python
st_19 = pd.read_csv('D:\John Sir\Pandas\Sales Transactions-2019.csv')
st_19.head()
```

## Left Merge

```python
ast18_19 = pd.merge(st_18,st_19, on='Product', how='left')
ast18_19
```

```python
wst18_19 = pd.merge(st_18,st_19, on='Date', how='left')
wst18_19
```

```python
left = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Shanu', 'Amar', 'Santosh', 'Rahul', 'Jayesh'],
   'Lang':['English','Hindi','Marathi','Telugu','Gujarathi']})
right = pd.DataFrame(
   {'id':[1,2,3,4,5],
   'Name': ['John', 'Dinesh', 'Ranbir', 'Rahul', 'Jayesh'],
   'Lang':['English','Hindi','Tamil','Bangali','Malyalam']})

print (pd.merge(left, right, on='Lang', how='left'))
```

## Right Merge

```python
bst18_19 = pd.merge(st_18,st_19, on='Product', how='right')
bst18_19
```

```python
xst18_19 = pd.merge(st_18,st_19, on='Date', how='right')
xst18_19
```

```python
left = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Shanu', 'Amar', 'Santosh', 'Rahul', 'Jayesh'],
   'Lang':['English','Hindi','Marathi','Telugu','Gujarathi']})
right = pd.DataFrame(
   {'id':[1,2,3,4,5],
   'Name': ['John', 'Dinesh', 'Ranbir', 'Rahul', 'Jayesh'],
   'Lang':['English','Hindi','Tamil','Bangali','Malyalam']})

print (pd.merge(left, right, on='Lang', how='right'))
```

## Outer Merge

```python
cst18_19 = pd.merge(st_18,st_19, on='Product', how='outer')
cst18_19
```

```python
yst18_19 = pd.merge(st_18,st_19, on='Date', how='outer')
yst18_19
```
